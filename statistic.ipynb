```python
dangvien = pd.read_csv(r'/Users/tuongvy/Documents/Tuong Vy/thư mục không có tiêu đề/myproject/dang_vien_phan_bo_tuoi_dang.csv')
dangvien.head(5)
```

```python
# thống kê số lượng đảng viên theo nhóm tuổi đảng, mỗi 10 năm là 1 nhóm (tuổi đảng là năm hiện tại trừ năm vào đảng), năm vào đảng lấy từ ngày vào đảng
import datetime
dangvien['NamVaoDang'] = pd.to_datetime(dangvien['NgayVaoDang'], errors='coerce').dt.year
dangvien['TuoiDang'] = datetime.datetime.now().year - dangvien['NamVaoDang']
dangvien['NhomTuoiDang'] = (dangvien['TuoiDang'] // 10) * 10
dangvien['NhomTuoiDang'] = dangvien['NhomTuoiDang'].astype(str) + '-' + (dangvien['NhomTuoiDang'].astype(int) + 9).astype(str)
dangvien['NhomTuoiDang'] = dangvien['NhomTuoiDang'].replace({'-9': '90+'})
nhom_tuoi_counts = dangvien['NhomTuoiDang'].value_counts().sort_index()
print(nhom_tuoi_counts)
```
